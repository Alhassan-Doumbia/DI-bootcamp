# What is Data Analysis?

Data analysis is the process of collecting, cleaning, transforming, and interpreting data in order to extract useful information and support decision-making. It helps individuals and organizations understand patterns, trends, and relationships within data.

The process usually involves several steps:
1. Collecting data
2. Cleaning and organizing data
3. Exploring and analyzing the data
4. Interpreting the results
5. Presenting insights through reports or visualizations



---

# Exercise 6 : Identifying Data Types

## Instructions

Below are various data sources. Identify whether each one is an example of **structured** or **unstructured** data.

---

# Questions

1. A company’s financial reports stored in an Excel file.  
2. Photographs uploaded to a social media platform.  
3. A collection of news articles on a website.  
4. Inventory data in a relational database.  
5. Recorded interviews from a market research study.

---

# Answers

| Data Source | Type of Data | Explanation |
|---|---|---|
| A company’s financial reports stored in an Excel file | **Structured Data** | The data is organized into rows and columns with a predefined format. |
| Photographs uploaded to a social media platform | **Unstructured Data** | Images do not follow a tabular structure and require special processing methods. |
| A collection of news articles on a website | **Unstructured Data** | Articles mainly contain free text without a fixed schema. |
| Inventory data in a relational database | **Structured Data** | Relational databases store data in organized tables with defined relationships. |
| Recorded interviews from a market research study | **Unstructured Data** | Audio recordings are not organized in a predefined tabular format. |

---

# Summary

- **Structured data** is highly organized and easy to store in tables or databases.
- **Unstructured data** does not follow a predefined format and often includes text, audio, images, or videos.


# 🌟 Exercise 7: Transformation Exercise

## Instructions

For each of the following unstructured data sources, propose a method to convert it into structured data and explain the reasoning.

---

# Answers

| Unstructured Data Source | Method to Convert into Structured Data | Reasoning |
|---|---|---|
| A series of blog posts about travel experiences | Use **Natural Language Processing (NLP)** to extract information such as destination, date, activities, and sentiment into a table or database. | Blog posts mainly contain free text. NLP techniques can identify important entities and organize them into structured columns. |
| Audio recordings of customer service calls | Apply **speech-to-text transcription** to convert audio into text, then categorize information such as customer issue, call duration, and satisfaction level. | Audio files are unstructured. Converting speech into text allows the data to be analyzed and stored in a structured format. |
| Handwritten notes from a brainstorming session | Use **Optical Character Recognition (OCR)** to digitize the handwritten text, then organize ideas into categories or keywords. | Handwritten notes cannot be directly analyzed by computers. OCR transforms them into machine-readable text that can be structured. |
| A video tutorial on cooking | Extract the audio using **speech recognition**, then identify steps, ingredients, and cooking times to create structured recipe data. | Videos contain visual and audio information. Speech recognition and data extraction help organize the content into a structured recipe format. |

---

# Summary

Unstructured data can be transformed into structured data by using technologies such as:

- **Natural Language Processing (NLP)** for text analysis
- **Speech-to-Text** for audio conversion
- **Optical Character Recognition (OCR)** for handwritten or printed text
- **Video and audio extraction techniques** for multimedia content

These methods help organize raw information into tables, databases, or datasets that are easier to analyze.
